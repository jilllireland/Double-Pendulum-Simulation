```
import sys
import numpy as np
#from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from sympy import *

# Pendulum rod lengths (m), bob masses (kg).
L1, L2 = 1., 5
m1, m2 = 1., 2
# The gravitational acceleration (m.s-2).
g = 9.81


Num_pts = 100000
t = np.linspace(0,20, num=Num_pts)

theta1dot = np.zeros(Num_pts)
theta11 = np.zeros(Num_pts)
theta12 = np.zeros(Num_pts)
theta13 = np.zeros(Num_pts)
theta14 = np.zeros(Num_pts)
theta15 = np.zeros(Num_pts)
theta2dot = np.zeros(Num_pts)
theta21 = np.zeros(Num_pts)
theta22 = np.zeros(Num_pts)
theta23 = np.zeros(Num_pts)
theta24 = np.zeros(Num_pts)
theta25 = np.zeros(Num_pts)
theta1ddot = np.zeros(Num_pts)
theta2ddot = np.zeros(Num_pts)
dt = t[1]-t[0]


def A(theta1,theta2,theta1dot,theta2dot):
    A = m1*L1**2 + m2*L1**2
    return A
def B(theta1,theta2,theta1dot,theta2dot):
    B = m2*L1*L2*np.cos(theta1-theta2)
    return B
def C(theta1,theta2,theta1dot,theta2dot):
    C = m2*L1*L2*np.sin(theta1-theta2)*theta2dot**2+(m1+m2)*g*L1*np.sin(theta1)
    return C
def D(theta1,theta2,theta1dot,theta2dot):
    D = m2*L2 ** 2
    return D
def E(theta1,theta2,theta1dot,theta2dot):
    E = m2*g*L2*np.sin(theta2)-m2*L1*L2*np.sin(theta1-theta2)*theta1dot**2
    return E


for i in range(Num_pts):
    if i==0:
        theta11[i]=30*np.pi/180
        theta21[i]=55*np.pi/180
        theta1dot[i]=np.pi/30
        theta2dot[i]=np.pi/30
    else:
        F = A(theta11[i-1],theta21[i-1],theta1dot[i-1],theta2dot[i-1])
        G = B(theta11[i-1],theta21[i-1],theta1dot[i-1],theta2dot[i-1])
        H = C(theta11[i-1],theta21[i-1],theta1dot[i-1],theta2dot[i-1])
        I = D(theta11[i-1],theta21[i-1],theta1dot[i-1],theta2dot[i-1])
        J = E(theta11[i-1],theta21[i-1],theta1dot[i-1],theta2dot[i-1])
        theta1ddot[i] = ((F*G*J - G**2*H) / ((F**2*I-F*G))) - (H/F)
        theta2ddot[i] = (G*H+F*J) / (I*F - G)
        theta1dot[i] = theta1dot[i-1]+theta1ddot[i-1]*dt
        theta2dot[i] = theta2dot[i-1]+theta2ddot[i-1]*dt
        theta11[i] = theta11[i-1]+theta1dot[i-1]*dt+0.5*g*dt**2
        theta21[i] = theta21[i-1]+theta2dot[i-1]*dt+0.5*g*dt**2
        
for i in range(Num_pts):
    if i==0:
        theta12[i]=35*np.pi/180
        theta22[i]=5*np.pi/180
        theta1dot[i]=np.pi/30
        theta2dot[i]=np.pi/30
    else:
        F = A(theta12[i-1],theta22[i-1],theta1dot[i-1],theta2dot[i-1])
        G = B(theta12[i-1],theta22[i-1],theta1dot[i-1],theta2dot[i-1])
        H = C(theta12[i-1],theta22[i-1],theta1dot[i-1],theta2dot[i-1])
        I = D(theta12[i-1],theta22[i-1],theta1dot[i-1],theta2dot[i-1])
        J = E(theta12[i-1],theta22[i-1],theta1dot[i-1],theta2dot[i-1])
        theta1ddot[i] = ((F*G*J - G**2*H) / ((F**2*I-F*G))) - (H/F)
        theta2ddot[i] = (G*H+F*J) / (I*F - G)
        theta1dot[i] = theta1dot[i-1]+theta1ddot[i-1]*dt
        theta2dot[i] = theta2dot[i-1]+theta2ddot[i-1]*dt
        theta12[i] = theta12[i-1]+theta1dot[i-1]*dt+0.5*g*dt**2
        theta22[i] = theta22[i-1]+theta2dot[i-1]*dt+0.5*g*dt**2
        
for i in range(Num_pts):
    if i==0:
        theta13[i]=40*np.pi/180
        theta23[i]=5*np.pi/180
        theta1dot[i]=np.pi/30
        theta2dot[i]=np.pi/30
    else:
        F = A(theta13[i-1],theta23[i-1],theta1dot[i-1],theta2dot[i-1])
        G = B(theta13[i-1],theta23[i-1],theta1dot[i-1],theta2dot[i-1])
        H = C(theta13[i-1],theta23[i-1],theta1dot[i-1],theta2dot[i-1])
        I = D(theta13[i-1],theta23[i-1],theta1dot[i-1],theta2dot[i-1])
        J = E(theta13[i-1],theta23[i-1],theta1dot[i-1],theta2dot[i-1])
        theta1ddot[i] = ((F*G*J - G**2*H) / ((F**2*I-F*G))) - (H/F)
        theta2ddot[i] = (G*H+F*J) / (I*F - G)
        theta1dot[i] = theta1dot[i-1]+theta1ddot[i-1]*dt
        theta2dot[i] = theta2dot[i-1]+theta2ddot[i-1]*dt
        theta13[i] = theta13[i-1]+theta1dot[i-1]*dt+0.5*g*dt**2
        theta23[i] = theta23[i-1]+theta2dot[i-1]*dt+0.5*g*dt**2

for i in range(Num_pts):
    if i==0:
        theta14[i]=45*np.pi/180
        theta24[i]=5*np.pi/180
        theta1dot[i]=np.pi/30
        theta2dot[i]=np.pi/30
    else:
        F = A(theta14[i-1],theta24[i-1],theta1dot[i-1],theta2dot[i-1])
        G = B(theta14[i-1],theta24[i-1],theta1dot[i-1],theta2dot[i-1])
        H = C(theta14[i-1],theta24[i-1],theta1dot[i-1],theta2dot[i-1])
        I = D(theta14[i-1],theta24[i-1],theta1dot[i-1],theta2dot[i-1])
        J = E(theta14[i-1],theta24[i-1],theta1dot[i-1],theta2dot[i-1])
        theta1ddot[i] = ((F*G*J - G**2*H) / ((F**2*I-F*G))) - (H/F)
        theta2ddot[i] = (G*H+F*J) / (I*F - G)
        theta1dot[i] = theta1dot[i-1]+theta1ddot[i-1]*dt
        theta2dot[i] = theta2dot[i-1]+theta2ddot[i-1]*dt
        theta14[i] = theta14[i-1]+theta1dot[i-1]*dt+0.5*g*dt**2
        theta24[i] = theta24[i-1]+theta2dot[i-1]*dt+0.5*g*dt**2      

for i in range(Num_pts):
    if i==0:
        theta15[i]=50*np.pi/180
        theta25[i]=5*np.pi/180
        theta1dot[i]=np.pi/30
        theta2dot[i]=np.pi/30
    else:
        F = A(theta15[i-1],theta25[i-1],theta1dot[i-1],theta2dot[i-1])
        G = B(theta15[i-1],theta25[i-1],theta1dot[i-1],theta2dot[i-1])
        H = C(theta15[i-1],theta25[i-1],theta1dot[i-1],theta2dot[i-1])
        I = D(theta15[i-1],theta25[i-1],theta1dot[i-1],theta2dot[i-1])
        J = E(theta15[i-1],theta25[i-1],theta1dot[i-1],theta2dot[i-1])
        theta1ddot[i] = ((F*G*J - G**2*H) / ((F**2*I-F*G))) - (H/F)
        theta2ddot[i] = (G*H+F*J) / (I*F - G)
        theta1dot[i] = theta1dot[i-1]+theta1ddot[i-1]*dt
        theta2dot[i] = theta2dot[i-1]+theta2ddot[i-1]*dt
        theta15[i] = theta15[i-1]+theta1dot[i-1]*dt+0.5*g*dt**2
        theta25[i] = theta25[i-1]+theta2dot[i-1]*dt+0.5*g*dt**2


# Size of figure in inches 
xSize = 10
ySize = 10
# Fontsize of various text elements of the figure in pts
fontSize = 14
# Size of line of the graph
lineSize = 3
# Color of line: 'r'=red, 'k'=black, 'b'=blue, 'g'=green, etc
lineColor = 'k'
# Tick labels size
tickSize = 12

#Plotting theta1 vs time:
fig1 = plt.figure(figsize=(xSize, ySize))
ax1 = fig1.add_subplot(111)
ax1.plot(t, theta11, color='k', linewidth=lineSize, label="theta1 = 5 degrees, theta2 = 5 degrees")
ax1.plot(t, theta12, color='red', linewidth=lineSize, label="theta1 = 10 degrees, theta2 = 5 degrees")
ax1.plot(t, theta13, color='blue', linewidth=lineSize, label="theta1 = 15 degrees, theta2 = 5 degrees")
ax1.plot(t, theta14, color='orange', linewidth=lineSize, label="theta1 = 20 degrees, theta2 = 5 degrees")
ax1.plot(t, theta15, color='green', linewidth=lineSize, label="theta1 = 25 degrees, theta2 = 5 degrees")
ax1.set_xlabel('Time (s)', fontsize=fontSize)
ax1.set_ylabel('theta1 (radians)', fontsize=fontSize)
ax1.tick_params(axis="x", labelsize=tickSize)
ax1.tick_params(axis="y", labelsize=tickSize)
ax1.legend(loc='upper right', fontsize=fontSize)

#Plotting theta2 vs time:
fig2 = plt.figure(figsize=(xSize, ySize))
ax2 = fig2.add_subplot(111)
ax2.plot(t, theta21, color='k', linewidth=lineSize, label="theta1 = 5 degrees, theta2 = 5 degrees")
ax2.plot(t, theta22, color='red', linewidth=lineSize, label="theta1 = 10 degrees, theta2 = 5 degrees")
ax2.plot(t, theta23, color='blue', linewidth=lineSize, label="theta1 = 15 degrees,theta2 = 5 degrees")
ax2.plot(t, theta24, color='orange', linewidth=lineSize, label="theta1 = 20 degrees, theta2 = 5 degrees")
ax2.plot(t, theta25, color='green', linewidth=lineSize, label="theta1 = 25 degrees, theta2 = 5 degrees")
ax2.set_xlabel('Time (s)', fontsize=fontSize)
ax2.set_ylabel('theta2 (radians)', fontsize=fontSize)
ax2.tick_params(axis="x", labelsize=tickSize)
ax2.tick_params(axis="y", labelsize=tickSize)
ax2.legend(loc='upper right', fontsize=fontSize)

#Plotting theta1 vs theta2:
fig2 = plt.figure(figsize=(xSize, ySize))
ax2 = fig2.add_subplot(111)
ax2.plot(theta21, theta11, color='k', linewidth=lineSize, label="theta1 = 5 degrees, theta2 = 5 degrees")
ax2.plot(theta22, theta12, color='red', linewidth=lineSize, label="theta1 = 10 degrees, theta2 = 5 degrees")
ax2.plot(theta23, theta13, color='blue', linewidth=lineSize, label="theta1 = 15 degrees, theta2 = 5 degrees")
ax2.plot(theta24, theta14, color='orange', linewidth=lineSize, label="theta1 = 20 degrees, theta2 = 5 degrees")
ax2.plot(theta25, theta15, color='green', linewidth=lineSize, label="theta1 = 25 degrees, theta2 = 5 degrees")
ax2.set_xlabel('theta1 (radians)', fontsize=fontSize)
ax2.set_ylabel('theta2 (radians)', fontsize=fontSize)
ax2.tick_params(axis="x", labelsize=tickSize)
ax2.tick_params(axis="y", labelsize=tickSize)
ax2.legend(loc='upper right', fontsize=fontSize)



n = t.size
Freq = np.fft.fftfreq(n, d=dt)
Freq = np.fft.fftshift(Freq)
theta11_F = np.fft.fft(theta11)
theta11_F = np.fft.fftshift(theta11_F)
theta12_F = np.fft.fft(theta12)
theta12_F = np.fft.fftshift(theta12_F)
theta13_F = np.fft.fft(theta13)
theta13_F = np.fft.fftshift(theta13_F)
theta14_F = np.fft.fft(theta14)
theta14_F = np.fft.fftshift(theta14_F)
theta15_F = np.fft.fft(theta15)
theta15_F = np.fft.fftshift(theta15_F)



# Create figure object
fig1 = plt.figure(figsize=(xSize, ySize))
# Create axes for Zoomed out real part
ax1 = fig1.add_subplot(221)
# Plot the real part of the frequency signal
ax1.plot(Freq, np.real(theta11_F), color='k', lw=lineSize, linewidth=lineSize, label="theta1 = 30 degrees")
ax1.plot(Freq, np.real(theta12_F), color='red', lw=lineSize, linewidth=lineSize, label="theta1 = 35 degrees")
ax1.plot(Freq, np.real(theta13_F), color='blue', lw=lineSize, linewidth=lineSize, label="theta1 = 40 degrees")
ax1.plot(Freq, np.real(theta14_F), color='orange', lw=lineSize, linewidth=lineSize, label="theta1 = 45 degrees")
ax1.plot(Freq, np.real(theta15_F), color='green', lw=lineSize, alpha=0.5, linewidth=lineSize, label="theta1 = 50 degrees")
# Create axes labels
ax1.set_xlabel('Frequency (Hz)', fontsize=fontSize)
ax1.set_ylabel('Frequency signal (a.u.)', fontsize=fontSize)
# Change size of tick labels
ax1.tick_params(axis="x", labelsize=tickSize)
ax1.tick_params(axis="y", labelsize=tickSize)
# Title
ax1.set_title('Real part', fontsize=fontSize)
#Zoom in
ax1.set_xlim(-1., 1.)
ax1.set_ylim(-160000, 100000)
#add legend
ax1.legend(bbox_to_anchor=(1.1, 1.05), loc='best')
# Show grid
plt.grid()

theta21_F = np.fft.fft(theta21)
theta21_F = np.fft.fftshift(theta21_F)
theta22_F = np.fft.fft(theta22)
theta22_F = np.fft.fftshift(theta22_F)
theta23_F = np.fft.fft(theta23)
theta23_F = np.fft.fftshift(theta23_F)
theta24_F = np.fft.fft(theta24)
theta24_F = np.fft.fftshift(theta24_F)
theta25_F = np.fft.fft(theta25)
theta25_F = np.fft.fftshift(theta25_F)

# Create figure object
fig2 = plt.figure(figsize=(xSize, ySize))
# Create axes for Zoomed out real part
ax2 = fig2.add_subplot(221)
# Plot the real part of the frequency signal
ax2.plot(Freq, np.real(theta21_F), color='k', lw=lineSize, linewidth=lineSize, label="theta1 = 5 degrees")
ax2.plot(Freq, np.real(theta22_F), color='red', lw=lineSize, linewidth=lineSize, label="theta1 = 10 degrees")
ax2.plot(Freq, np.real(theta23_F), color='blue', lw=lineSize, linewidth=lineSize, label="theta1 = 15 degrees")
ax2.plot(Freq, np.real(theta24_F), color='orange', lw=lineSize, linewidth=lineSize, label="theta1 = 20 degrees")
ax2.plot(Freq, np.real(theta25_F), color='green', lw=lineSize, alpha=0.5, linewidth=lineSize, label="theta1 = 25 degrees")
# Create axes labels
ax2.set_xlabel('Frequency (Hz)', fontsize=fontSize)
ax2.set_ylabel('Frequency signal (a.u.)', fontsize=fontSize)
# Change size of tick labels
ax2.tick_params(axis="x", labelsize=tickSize)
ax2.tick_params(axis="y", labelsize=tickSize)
# Title
ax2.set_title('Real part', fontsize=fontSize)
#Zoom in
ax2.set_xlim(-1., 1.)
ax2.set_ylim(-160000, 100000)
#add legend
ax2.legend(bbox_to_anchor=(1.1, 1.05), loc='best')
# Show grid
plt.grid()


# Create figure object
fig2 = plt.figure(figsize=(xSize, ySize))
# Create axes for Zoomed out real part
ax2 = fig2.add_subplot(221)
# Plot the real part of the frequency signal
ax2.plot(Freq, np.real(theta21_F), color='k', lw=lineSize, linewidth=lineSize, label="theta1 = 30 degrees theta 2 = 50 degrees")
# Create axes labels
ax2.set_xlabel('Frequency (Hz)', fontsize=fontSize)
ax2.set_ylabel('Frequency signal (a.u.)', fontsize=fontSize)
# Change size of tick labels
ax2.tick_params(axis="x", labelsize=tickSize)
ax2.tick_params(axis="y", labelsize=tickSize)
# Title
ax2.set_title('Real part', fontsize=fontSize)
#Zoom in
ax2.set_xlim(-1., 1.)
ax2.set_ylim(-160000, 100000)
#add legend
ax2.legend(bbox_to_anchor=(1.1, 1.05), loc='best')
# Show grid
plt.grid()
```
